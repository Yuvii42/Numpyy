## Matplot Library

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([50, 40, 70, 80])
y2 = np.array([5, 15, 20, 30])
y3 = np.array([30, 25, 40, 50])

line_style = dict(marker=".", markersize=10, linestyle="dashdot")

plt.plot(x,y1, color = "#169d23", **line_style)
plt.plot(x,y2,color = "#144fdc", **line_style)
plt.plot(x, y3, **line_style)
plt.title("Year vs Value")
plt.show()
```

### Labels: 

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([50, 40, 70, 80])
y2 = np.array([5, 15, 20, 30])
y3 = np.array([30, 60, 40, 50])

line_style = dict(marker=".", markersize=10, linestyle="dashdot")
plt.title("Year vs Class Stength", fontsize=14, family="Arial", fontweight="bold", color="#2d4cfc")

plt.xlabel("Year", fontsize=12, family="Arial", fontweight="bold", color="#fc3c2d")
plt.ylabel("Class Strength", fontsize=12, family="Arial", fontweight="bold", color="#fc3c2d")

plt.tick_params(axis='both', which='major', labelsize=10, color="#34fc2d")

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

### Grid Lines: Helps make plot easier to read by adding reference lines.


```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

plt.grid(axis='both', linewidth=2, color='lightgray', linestyle="dashed")
plt.plot(x,y)
plt.show()
```

### Bar Chart: Compare a category of data represent in each category

```python
categories = np.array(['Grains', 'Fruits', 'Vegies', 'Protien', 'Dairy', 'Sweets'])
values = np.array([4, 3, 2, 5, 3, 1])

# plt.bar(categories, values, color="gray")
plt.barh(categories, values, color="gray")
plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")
plt.show()
```

### Pie Chart: Circular chart divided into slices to show percentages of the total.
###            Good for Visualizing distribution amnong categories

```python
categories = ["Freshman", "Sophomore", "Junior", "Senior"]
values = np.array([120, 100, 80, 60])

colors = ["#a93c3c","#023970","#000000","#884f15"]
plt.pie(values, labels=categories, autopct="%1.1f%%", startangle=90, colors=colors, explode=[0,0,0,0.1], shadow=True)
plt.title("Class Distribution")
plt.show()
```

### Scatter Graph:  Shows a relationship between two variable and Helps to identify a correlation(+, -, None). For example: Study Hour vs Test Score

```python
x1 = np.array([0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8])
y1 = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87])

x2 = np.array([0, 1, 2, 2, 3, 4,5,6,7,8,8])
y2 = np.array([55, 58, 65, 70, 68, 72, 78, 78, 83, 92, 95])

plt.scatter(x1, y1, color="purple", marker=".", s=200, label="Class A")
plt.scatter(x2, y2, color="red", marker=".", s=200, label="Class B")

plt.xlabel("Hours Studied")
plt.ylabel("Grade")
plt.title("Test Score")
plt.legend()
plt.show()
```

### Histogram:  A visual Representation of the distribution of q uantitive data. They group values in bins (intervals) and count how many fall in each range.

```python
score = np.random.normal(loc=80, scale=10, size=100)
score = np.clip(score,0,100)

plt.hist(score, bins=10, color='lightgreen', edgecolor='black')
plt.title("Exam Score Distribution")
plt.xlabel("Score")
plt.ylabel("Number of Students")

plt.show()
```

### Subplots:  # Figure = The entire Canvas and Ax = A single plot (subplot)

```python
x = np.array([1,2,3,4,5])

figure, axes = plt.subplots(2,2)

axes[0,0].plot(x, x*2, color='red')
axes[0,0].set_title("x*2")

axes[0,1].bar(x, x**2, color='green')
axes[0,1].set_title("x**2")

axes[1,0].plot(x, x**3, color='blue')
axes[1,0].set_title("x**3")

axes[1,1].plot(x, x**4, color='yellow')
axes[1,1].set_title("x**4")

plt.tight_layout()
plt.show()
```
