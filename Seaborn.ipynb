## Seaborn it already have number of inbuild data set.

```python
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
penguins['island'].value_counts()
penguins['species'].value_counts()
```

```python
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue="island")
```

## The way of Styling themes are as follow:

```python
sns.set_style("ticks")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue="island", palette="Dark2")
sns.despine(left=True)
```

```python
sns.scatterplot(data=penguins, x="species", y="body_mass_g", hue = "island", style="sex" )
```

```python
sns.stripplot(data=penguins, x = "species",  y="body_mass_g", hue = "island", jitter=True, dodge=True)
```

```python
#Swarm Plot
sns.set_context("talk")
sns.swarmplot(data = penguins, x="species", y="body_mass_g", hue="island")
```

```python
# Histogram
sns.set_context("notebook")
sns.histplot(data=penguins, x = "body_mass_g", hue="sex", multiple="stack")
```

```python
# Regression Plot
sns.regplot(data=penguins, x = "body_mass_g", y="flipper_length_mm", color="green")
```

```python
#line Plot
sns.set_style("whitegrid")
sns.lineplot(data=penguins, y="flipper_length_mm", x="body_mass_g", hue="island", style="sex")
```

```python
#Joint Plot
sns.jointplot(data = penguins, x = "body_mass_g", y="flipper_length_mm")
```

```python
sns.jointplot(data=penguins, x = "body_mass_g", y = "flipper_length_mm", hue ="sex")
```

```python
sns.jointplot(data=penguins, x = "body_mass_g", y = "flipper_length_mm", hue ="sex", kind="kde")
```

```python
sns.jointplot(data=penguins, x = "body_mass_g", y = "flipper_length_mm", hue ="island", kind="scatter")
```

```python
#bar plot
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette=["red", "blue"], estimator=np.var)
```

```python
#Countplot

sns.countplot(data=penguins, x="species", hue="island")
```

```python
#Box plot

sns.boxplot(data=penguins, x="species", y="body_mass_g", hue="sex")
```

```python
#Violin Plot
sns.violinplot(data=penguins, x="species", y="body_mass_g", hue="sex", split=True)
```

```python
sns.violinplot(data=penguins, x="species", y="body_mass_g", palette="Pastel1")
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue="island", color="k", alpha=0.5)
```

### KDE -> For estimating the probabiltiy density function of continous or non parametric data, or in simp,e terms: a smooth curve that shows the distribution of data like  a smoothed. It helps you to see where the data is concentrated (peaks) and how it spreads.

```python
sns.kdeplot(data=penguins, x = "body_mass_g", hue="species", palette="pastel", fill=False)
sns.despine()
```

```python
#Heatmap: a 2D color coded matrix that shows relationship or pattern between two variables.
penguins.head()
```

```python
column = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
penguins[column].corr()
sns.heatmap(penguins[column].corr(), annot=True, cmap="coolwarm", vmin=-0.2, linewidths="2", linecolor="black")
```

```python
# Rug Plot
sns.rugplot(data=penguins, x="body_mass_g", hue="species", palette="Set2", height=0.2)
```

```python
# pair plot

sns.pairplot(data=penguins, hue = "sex", palette="Set2", diag_kind="kde")
```

```python
# Pair grid -> subplot grid for plotting pairwise relationship in a dadaset. It allows you to plot multiple bivariate relationships in a dataset.

graph = sns.PairGrid(data=penguins, hue="sex", palette="Set2")
graph.map_upper(sns.scatterplot) #upper triangle 
graph.map_lower(sns.kdeplot)    #lower Triangle
graph.map_diag(sns.histplot, kde=True)  #Diagonal Triangle
graph.add_legend()
```
