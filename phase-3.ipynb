## Advance operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure: [restaurent_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #Paradise Biryani
    [2, 120000, 140000, 160000, 190000], #Beijing Bites 
    [3, 200000, 230000, 260000, 300000], #Pizza Hub
    [4, 180000, 210000, 240000, 270000], #Burger Point
    [4, 160000, 185000, 205000, 230000], #Chai Point
]) 

print("====== Zomato Sales Analysis ======")
print("\n Sales date shape", sales_data.shape)
print(" \n Sample data for 1st 3 result: \n", sales_data[0:3])
```

```python
#total sales for each year
print(np.sum(sales_data, axis=0))

#Yearly Sales
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)

#minimum sales per year
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Minimum sales per restaurent:", min_sales)

#maximum sales per restaurent
max_sales = np.max(sales_data[:, 1:], axis=0)
print("Maximum sales per restaurent:", max_sales)


#Average sales per restaurent
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average sales per restaurent:", avg_sales)

```

```python
#cumulative sales over the years
cumulative_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative sales over the years:\n", cumulative_sales)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumulative_sales, axis=0))
plt.title("Average Cumulative Sales Over Years")
plt.xlabel("Years (2021-2024)")
plt.ylabel("Average Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([10,20,30,40,50])

print("Vector Addition:", vector1 + vector2)
print("Multiplication:", vector1 * vector2)
print("Dot Product:", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```

```python
restaurant_types = np.array(['biryani', 'chinese', 'pizza', 'burger', 'cafe'])
vectorized_upper = np.vectorize(str.upper)
print("Vectorized Upper", vectorized_upper(restaurant_types))
```
